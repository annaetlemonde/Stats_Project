```python
df['Close_ETF'].mean()
```

### Case I: divide into 20 groups with 50 in each group sequentially 

```python
df['group'] = pd.cut(df['Close_ETF'],bins = 20, labels = False)
```

```python
mean = []
for i in range(20):
    mu = df[df['group'] == i]['Close_ETF'].mean()
    mean.append(mu)
```

```python
df_mean = pd.DataFrame({'Mean':mean})
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
df_mean['Mean'].hist()
plt.title('Histogram for Sample Mean with sample 50')
```

The distribution is a skewed distribution and we have no specific name for this type of distribution, and it's definitely not normal. 

```python
df_mean['Mean'].mean()
```

```python
df_mean['Mean'].std()
```

```python
df['Close_ETF'].std()/math.sqrt(20)
```

The sample mean and sample standard deviation both are larger than the population mean and population standard deviation. 

Result from 3 and 4 are not consistent, because the sample is sequential, not random. 

### Case II: divide into 10 groups with 100 in each group sequentially 

```python
df['group2'] = pd.cut(df['Close_ETF'],bins = 10, labels = False)
```

```python
mean2 = []
for i in range(10):
    mu2 = df[df['group'] == i]['Close_ETF'].mean()
    mean2.append(mu2)
```

```python
df_mean2 = pd.DataFrame({'Mean':mean2})
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
df_mean2['Mean'].hist()
plt.title('Histogram for Sample Mean with sample 100')
```

According to the histogram, the distribution is definitely not normal. 

```python
df_mean2['Mean'].mean()
```

```python
df_mean2['Mean'].std()
```

```python
df['Close_ETF'].std()/math.sqrt(10)
```

Sample mean is smaller than population mean and sample standard deviation is larger than population standard deviation. 

Result from 3 and 4 are not consistent, because the group is sequential, not random. 

### Case III: divide into 50 groups with 20 in each group randomly 

```python
mean3 = []
for i in range(50):
    a = np.random.choice(df.Close_ETF, size = 20, replace = True).mean()
    mean3.append(a)
```

```python
df_mean3 = pd.DataFrame({'Mean':mean3})
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
df_mean3['Mean'].hist()
plt.title('Histogram for Sample Mean with sample 20')
```

The histogram is close to normal, but it's not normal. 

```python
df_mean3['Mean'].mean()
```

```python
df_mean3['Mean'].std()
```

```python
df['Close_ETF'].std()/math.sqrt(50)
```

Sample mean is really close to the population mean, and sample standard deviation is greater than population mean. 

Both values are really close, but not consistent with CLT, because each sample size is too small.
